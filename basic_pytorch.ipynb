```
# load modules
import numpy as np
import matplotlib.pyplot as plt

# create toy dataset
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]


# our model for the forward pass
# inputs times weights
def forward(x):
    return x * w


# loss function
# in this case, mean square Error (MSE)
def loss(x, y):
    y_pred = forward(x)
    return (y_pred - y) * (y_pred - y)

# List of weights/MSE for each input
w_list = []
mse_list = []

for w in np.arange(0.0, 4.1, 0.1):
    # print the weights and initialize the lost
    print("w=", w)
    l_sum = 0

    for x_val, y_val in zip(x_data, y_data):
        # for each input and output, calculate y_hat
        # compute the total loss and add to the total error
        y_pred_val = forward(x_val)
        l = loss(x_val, y_val)
        l_sum += l
        print("\t", x_val, y_val, y_pred_val, l)
    # compute the MSE of each
    # append the weight/mse from this run
    print("MSE=", l_sum / 3)
    w_list.append(w)
    mse_list.append(l_sum / 3)

# plot results
plt.plot(w_list, mse_list)
plt.ylabel('Loss')
plt.xlabel('w')
plt.show()
```
